```python
import pandas as pd
import sqlite3
import folium
```

# Accidents in São Paulo

```python
con = sqlite3.connect('../data/mydb')
```

```python
alerts = pd.read_sql_query('select * from alerts where city = "São Paulo"', con)
```

```python
alerts.head(1)
```

```python
print(alerts.groupby('type').count()['city']['ACCIDENT'], "accidents where reported in São Paulo in the last 5 minutes" )
```

```python
acc = alerts[alerts['type'] == 'ACCIDENT']
```

```python
acc.columns
```

```python
swap = lambda x: [(i[1], i[0]) for i in x]
```

```python
def plot_map(polygons):

    m = folium.Map(
        location=polygons[0],
        zoom_start=12
    )
    for poly in polygons:

        folium.Marker(poly).add_to(m)
    return m
```

```python
plot_map(swap(list(map(list, list(acc[['location.x', 'location.y']].values)))))
```
